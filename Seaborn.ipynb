```python
# importing seaborn library
import seaborn as sns
```

```python
# loading dataset
tips = sns.load_dataset("tips")
tips
```

```python
tips.head(5)
```

```python
## Scatter Plot 
import matplotlib.pyplot as plt 
sns.scatterplot(x = "total_bill" , y = "tip" , data = tips)
plt.title("Scatter Plot of total_bill vs tip")
plt.show()
```

```python
sns.scatterplot(x = "tip" , y = "size" , data = tips)
plt.show()
```

```python
## Line Plot 
sns.lineplot(x = "size" , y = "total_bill" , data =tips)
plt.title("Line pot")
plt.show()
```

```python
sns.lineplot(x = "size" , y = "tip" , data = tips)
plt.show()
```

```python
## Categorical Plots

# Bar Plot
sns.barplot(x = "day" , y = "total_bill" , data = tips)
plt.title("Bar plot of total bills per/by day")
plt.show()
```

```python
sns.barplot(x = "sex" , y = "tip" , data = tips)
plt.show()
```

```python
## Box Plot- It gives 0 , 25 , 50 , 75 , 100 percentile of data and also give outliers 
sns.boxplot(x = "day" , y = "total_bill" , data = tips )
plt.show()
```

```python
sns.boxplot(x = "sex" , y = "tip" , data = tips)
plt.show()
```

```python
# Violin plot - for distribution of data 
sns.violinplot(x = "day" , y = "total_bill" , data = tips)
plt.show()
```

```python
sns.violinplot(x = "sex" , y ="size" , data = tips)
plt.show()
```

```python
## histogram 
sns.histplot(tips["total_bill"] , bins = 10 , kde = True)
```

```python
## kde ( kernel density estimator ) plot
sns.kdeplot(tips["total_bill"] , shade = True , fill = True)
plt.show()
```

```python
sns.histplot(tips["tip"],bins = 5 ,kde=True)
plt.show()
```

```python
## pair plot - relationship between all the variables ( only numerical data )
sns.pairplot(tips)
```

```python
# Heat Map
corr = tips[["total_bill", "tip" , "size"]].corr()
corr
```

```python
## change in tip is 0.675 with respect to 1.00 unit total_bill and so on 
```

```python
sns.heatmap(corr,annot = True)
```

```python
sns.heatmap(corr,annot = True , cmap= "coolwarm")
```
