```python
## merging and joining DataFrames
df1 = pd.DataFrame({"Key" : ["A","B","C"] , "Value1" : [1,2,3]})
df2 = pd.DataFrame({"Key" : ["A","B","D"] , "Value2" : [4,5,6]})
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2)
```

```python
pd.merge(df1,df2, on = "Key" , how = "inner") # on = column name and how = type of join 
```

```python
pd.merge(df1,df2,on = "Key" , how = "outer")
```

```python
pd.merge(df1,df2,on ="Key",how ="left")
```

```python
pd.merge(df1,df2,on ="Key",how = "right")
```

```python
# joins 
# inner join = common values only 
# outer join = all the values 
# left outer join = left df is given priority
# right outer join = right df is given priority
```
